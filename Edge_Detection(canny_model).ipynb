this code will dynamically access files and fetch the iamge


```python
from google.colab import files
import cv2
from matplotlib import pyplot as plt

uploaded = files.upload()
filename = next(iter(uploaded))
image = cv2.imread(filename)
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

convert image to grayscale using openCV libaray


```python
imag = cv2.imread(filename,0)
# cmap should be explicitly defined gray
plt.imshow(imag, cmap = 'gray')
plt.axis('off')
plt.show()
```

Removing the noise from the grayscaled image


```python
blur_image = cv2.GaussianBlur(imag,(5,5),0)
plt.imshow(blur_image, cmap = 'gray')
plt.axis('off')
plt.show()
```

using canny model for edge detection


```python
t_lower = 50
t_upper = 50
canny_image = cv2.Canny(blur_image,t_lower,t_upper)
plt.imshow(canny_image, cmap = 'gray')
plt.axis('off')
plt.show()
```
